```python
plt.figure( figsize= ( 12,4) )
plt.plot(x ,X[:, mass_index], 'y-', label = 'Indice de masa', alpha = 0.5)
plt.plot(x ,X[:, blood_pressure], 'b-', label = 'Presión sanguinea', alpha = 0.5)
plt.legend(prop= {'size': 9})
plt.grid()
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize= (12,4))

# Diagramas de dispersión para varias caraterísticas

for fig, feature, label in zip(range(2), [mass_index, age], ['Indice de masa', 'Edad']):
    axs[fig].scatter(X[:,feature], X[:, blood_pressure], c='g', marker='o', s= 5)
    axs[fig].set_xlabel(label)
    axs[fig].set_ylabel('Presión arterial')
    axs[fig].grid()

plt.show()
```
